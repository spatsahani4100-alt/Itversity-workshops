```python
from pyspark.sql import SparkSession
```

```python
session = SparkSession. \
    builder. \
    appName("Car Sales Analysis"). \
    getOrCreate()
```

```python
car_sales = session. \
    read. \
    csv('car_sales_2024.csv', 
        header=True, 
        inferSchema=True
    )
```

```python
car_sales.count()
```

```python
car_sales.show()
```

```python
# Get the sales for a given state
state = input("Enter the state: ")
state_sales = car_sales.filter(f"state == '{state}'")
```

```python
state_sales.count()
```

```python
state_sales.show()
```

```python
state_sales.select("sale_id", "sale_date", "state").show()
```

```python
from pyspark.sql.functions import sum, count, round, col
```

```python
# Get sales by state
sales_by_state = car_sales. \
    groupBy('state'). \
    agg(
        round(sum('sale_price'), 2).alias('total_sales'), \
        count('sale_price').alias('sales_count')
    )
```

```python
sales_by_state.count()
```

```python
sales_by_state.show()
```

```python
# Convert to Pandas for better formatting
sales_by_state.toPandas()
```

```python
sales_by_state. \
    orderBy(col('total_sales').desc()). \
    toPandas()
```
