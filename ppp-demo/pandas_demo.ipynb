```python
import pandas as pd
```

```python
car_sales = pd.read_csv('car_sales_2024.csv')
```

```python
car_sales.shape
```

```python
car_sales.head()
```

```python
car_sales.columns
```

```python
# Get the sales for a given state
state = input("Enter the state: ")
state_sales = car_sales.query("state == @state")
```

```python
state_sales.shape
```

```python
state_sales.head()
```

```python
state_sales[["sale_id", "sale_date", "state"]]
```

```python
# Get sales by state
state_sales = car_sales.groupby('state'). \
    agg(total_sales=('sale_price', 'sum'), \
        sale_count=('sale_price', 'count'))
```

```python
state_sales.shape
```

```python
state_sales.head()
```

```python
state_sales. \
    reset_index(). \
    sort_values('total_sales', ascending=False)
```
