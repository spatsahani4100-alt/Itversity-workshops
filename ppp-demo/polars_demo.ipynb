```python
import polars as pl
```

```python
car_sales = pl.read_csv('car_sales_2024.csv')
```

```python
car_sales.shape
```

```python
car_sales.head()
```

```python
# Get the sales for a given state
state = input("Enter the state: ")
state_sales = car_sales.filter(pl.col('state') == state)
```

```python
state_sales.shape
```

```python
state_sales.head()
```

```python
state_sales.select("sale_id", "sale_date", "state")
```

```python
# Get sales by state using polars
sales_by_state = car_sales.group_by('state').agg(
    pl.sum('sale_price').alias('total_sales'),
    pl.count('sale_price').alias('sales_count')
)
```

```python
sales_by_state.shape
```

```python
sales_by_state.head()
```

```python
# Configure Polars to display numbers without scientific notation
pl.Config.set_fmt_float("full")
```

```python
sales_by_state.head()
```

```python
sales_by_state.sort('total_sales', descending=True)
```
